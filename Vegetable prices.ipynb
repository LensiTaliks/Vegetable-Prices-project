```python
import pandas as pd

# Define the file path (using raw string to avoid issues with backslashes)
file_path = r"C:\Users\obake\Downloads\archive\prices.csv"

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first 5 rows of the dataset
df.head()

```

```python
# Check for missing values in the dataset
df.isnull().sum()
```

```python
# Drop rows with any missing values
df_cleaned = df.dropna()

# Drop columns with missing values
df_cleaned = df.dropna(axis=1)
```

```python
# Drop rows with any missing values
df = df.dropna()

# Or drop columns with missing values
df = df.dropna(axis=1)
```

```python
# Convert the 'Price Dates' column to datetime
df['Price Dates'] = pd.to_datetime(df['Price Dates'], format='%d-%m-%Y')

# Verify the conversion
print(df['Price Dates'].dtype)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply standardization to all numeric columns (except 'Price Dates')
df_scaled = df.copy()
df_scaled[['Bhindi (Ladies finger)', 'Tomato', 'Onion', 'Potato', 'Brinjal', 'Garlic', 'Peas', 'Methi', 'Green Chilli', 'Elephant Yam (Suran)']] = scaler.fit_transform(
    df[['Bhindi (Ladies finger)', 'Tomato', 'Onion', 'Potato', 'Brinjal', 'Garlic', 'Peas', 'Methi', 'Green Chilli', 'Elephant Yam (Suran)']]
)

# Verify the scaled DataFrame
df_scaled.head()
```

```python
from scipy import stats

# Calculate Z-scores for all numeric columns
z_scores = stats.zscore(df[['Bhindi (Ladies finger)', 'Tomato', 'Onion', 'Potato', 'Brinjal', 'Garlic', 'Peas', 'Methi', 'Green Chilli', 'Elephant Yam (Suran)']])

# Define a threshold for outliers (usually 3)
threshold = 3
df_no_outliers = df[(abs(z_scores) < threshold).all(axis=1)]

# Verify the DataFrame without outliers
df_no_outliers.head()
```
